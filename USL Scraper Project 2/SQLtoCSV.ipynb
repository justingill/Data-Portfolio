This file is used to transform our SQL table to a CSV because Tableau does not support the current version of sqlite3 unfortunately.

```python
import pandas as pd
import sqlite3
import datetime as datetime
```

```python
con = sqlite3.connect('USL.sqlite')
usl= pd.read_sql_query('Select * from "{}"'.format(str(datetime.date.today())),
                             con,
                             coerce_float=True)
```
